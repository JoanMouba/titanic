# Importing the needed libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Load Titanic data and explore statistics

```python
passengers = pd.read_csv("titanic_train.csv")
col_headers = passengers.iloc[0]

print(col_headers)
print(passengers.head())
print(passengers.tail())
```

# Basics statistics

```python
passengers.info()
```

```python
passengers.describe()
```

```python
cols =["Survived","Pclass", "Name", "Sex", "Age", "SibSp", "Parch", "Ticket", "Fare", "Embarked"]
```

```python
# Drop columns : PassengerId and Cabin 
passengers_with_dropped_cols = passengers[cols]
```

```python
passengers_with_dropped_cols.info()
```

```python
passengers_cleaned = passengers_with_dropped_cols.dropna()
```

```python
passengers_cleaned.info()
```

```python
shape_passengers = passengers.shape
shape_passengers_cleaned = passengers_cleaned.shape
raw_data_count = shape_passengers[0]
cleaned_data_count = shape_passengers_cleaned[0]
print(shape_passengers)
print(shape_passengers_cleaned)
print("raw_data_count: ", raw_data_count)
print("cleaned_data_count: ", cleaned_data_count)
```
